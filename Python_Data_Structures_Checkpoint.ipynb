```python
# Create a list to store the shopping items
shopping_list = []

# Define a set to store unique items
unique_items = set()

# Define a dictionary to store item quantities
item_quantities = {}

# Define the maximum number of items that can be added
max_items = 10

# Function to display the menu
def display_menu():
    print("\nMenu:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View items")
    print("4. Exit")

# Function to add an item
def add_item():
    if len(shopping_list) < max_items:
        item = input("Enter the item to add: ")
        shopping_list.append(item)
        unique_items.add(item)
        if item in item_quantities:
            item_quantities[item] += 1
        else:
            item_quantities[item] = 1
        print(f"'{item}' has been added to the shopping list.")
    else:
        print("The shopping list is full. You cannot add more items.")

# Function to remove an item
def remove_item():
    item = input("Enter the item to remove: ")
    if item in shopping_list:
        shopping_list.remove(item)
        if item_quantities[item] > 1:
            item_quantities[item] -= 1
        else:
            del item_quantities[item]
            unique_items.remove(item)
        print(f"'{item}' has been removed from the shopping list.")
    else:
        print(f"'{item}' is not in the shopping list.")

# Function to view items
def view_items():
    if shopping_list:
        print("\nShopping List:")
        for item in shopping_list:
            print(f"- {item} (Quantity: {item_quantities[item]})")
    else:
        print("The shopping list is empty.")

# Main loop
while True:
    display_menu()
    choice = input("Enter your choice: ")
    if choice == '1':
        add_item()
    elif choice == '2':
        remove_item()
    elif choice == '3':
        view_items()
    elif choice == '4':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```
